```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np
import json


def analyze_interactions(female_files, male_files, network_measure_func, measure, interaction):
    
    d_female = {}
    d_female_measure = {}
    for i in range(len(female_files)):
        d_female["group" + str(i) + "_female"] = nx.read_gml(female_files[i])
        d_female_measure["group" + str(i) + "_female"] = network_measure_func(d_female["group" + str(i) + "_female"])
    
    
    total_dict_female = Counter()
    for d in d_female_measure.values():
        total_dict_female.update(d)
    total_dict_female = dict(total_dict_female)
    print("female " + str(total_dict_female))
    
  
    
    d_male = {}
    d_male_measure = {}
    for i in range(len(male_files)):
        d_male["group" + str(i) + "_male"] = nx.read_gml(male_files[i])
        d_male_measure["group" + str(i) + "_male"] = network_measure_func(d_male["group" + str(i) + "_male"])
      
    total_dict_male = Counter()
    for d in d_male_measure.values():
        total_dict_male.update(d)
    total_dict_male = dict(total_dict_male)
    print("male " + str(total_dict_male))
    
    names = list(total_dict_male.keys())
    values_male = list(total_dict_male.values())
    values_female = list(total_dict_female.values())
    
    N = len(female_files)*2
    ind = np.arange(N) 
    width = 0.35       
    plt.bar(ind, values_male, width, label='Male')
    plt.bar(ind + width, values_female, width, label='Female')

    plt.ylabel('Values')
    plt.title(str(measure) + ' in ' + str(interaction) + " by Group and Gender")

    plt.xticks(ind + width / 2, ('1-CC', '2-CC', '3-CC', '1-FC', '2-FC', '3-FC'))
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.savefig(str(measure.split(" ")[0]) + "_" + str(measure.split(" ")[1]) + "_" + str(interaction) + "_bar.png", bbox_inches='tight')
    plt.show()
    
    
```

# Combine FC and CC

```python
def combine_analyze_interactions(female_files, male_files, network_measure_func, measure, interaction):
    d_female = {}
    d_female_measure = {}
    for i in range(len(female_files)):
        d_female["group" + str(i) + "_female"] = nx.read_gml(female_files[i])
        d_female_measure["group" + str(i) + "_female"] = network_measure_func(d_female["group" + str(i) + "_female"])
    
    total_dict_female = Counter()
    for d in d_female_measure.values():
        total_dict_female.update(d)
    total_dict_female = dict(total_dict_female)
    print(total_dict_female)
    
    new_dict_female = {'CC': 0, 'FC': 0} 
    for i in total_dict_female.items():
        if i[0].endswith("CC"):
            new_dict_female['CC'] += i[1]
        else:
            new_dict_female['FC'] += i[1]    
    print(new_dict_female)
    print("======================")
    d_male = {}
    d_male_measure = {}
    for i in range(len(male_files)):
        d_male["group" + str(i) + "_male"] = nx.read_gml(male_files[i])
        d_male_measure["group" + str(i) + "_male"] = network_measure_func(d_male["group" + str(i) + "_male"])
    
    total_dict_male = Counter()
    for d in d_male_measure.values():
        total_dict_male.update(d)
    total_dict_male = dict(total_dict_male)
    print(total_dict_male)
    new_dict_male = {'CC': 0, 'FC': 0} 
    for i in total_dict_male.items():
        if i[0].endswith("CC"):
            new_dict_male['CC'] += i[1]
        else:
            new_dict_male['FC'] += i[1]
    print(new_dict_male)
    names = list(new_dict_male.keys())
    values_male = list(new_dict_male.values())
    values_female = list(new_dict_female.values())
    
    N = 2
    ind = np.arange(N) 
    width = 0.35       
    plt.bar(ind, values_male, width, label='Male')
    plt.bar(ind + width, values_female, width, label='Female')

    plt.ylabel('Values')
    plt.title(str(measure) + ' in ' + str(interaction) + " by Group and Gender")

    plt.xticks(ind + width / 2, ('CC', 'FC'))
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.savefig("combined_" + str(measure.split(" ")[0]) + "_" + str(measure.split(" ")[1]) + "_" + str(interaction) + "_bar.png", bbox_inches='tight')
    plt.show()
    
    
```
